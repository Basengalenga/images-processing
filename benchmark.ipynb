```python
import sys
import os
from PIL import Image
import matplotlib.pyplot as plt
import time
```

```python
sys.path.append('python')

from gaussian import apply_gaussian
from sobel import apply_sobel
from median import apply_median

# Input
INPUT_PATH = 'data/epic_imagen.jpg'

# Aplica los 3 filtros
apply_gaussian(INPUT_PATH, 'data/gaussian_output.png')
apply_sobel(INPUT_PATH,    'data/sobel_output.png')
apply_median(INPUT_PATH,   'data/median_output.png')

# Visualiza
images = {
    'Original': INPUT_PATH,
    'Gaussian': 'data/gaussian_output.png',
    'Sobel':    'data/sobel_output.png',
    'Median':   'data/median_output.png',
}

fig, axes = plt.subplots(1, 4, figsize=(18, 5))

for ax, (title, path) in zip(axes, images.items()):
    img = Image.open(path)
    ax.imshow(img, cmap='gray')
    ax.set_title(title, fontsize=13)
    ax.axis('off')

plt.tight_layout()
plt.show()
print("proceso terminado")
```
