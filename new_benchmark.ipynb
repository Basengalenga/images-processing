```python
import sys
import os
import time
from PIL import Image
import matplotlib.pyplot as plt
```

```python
sys.path.append('python')

from gaussian import apply_gaussian
from sobel import apply_sobel
from median import apply_median

# Input
INPUT_PATH = 'data/epic_imagen.jpg'

# Diccionario para guardar los tiempos
tiempos = {}

def medir_tiempo(nombre, funcion, *args):
    inicio = time.time()
    funcion(*args)
    fin = time.time()
    duracion = fin - inicio
    tiempos[nombre] = duracion
    print(f"✅ {nombre} completado en: {duracion:.4f} segundos")

# --- Ejecución y Medición ---
medir_tiempo('Gaussian', apply_gaussian, INPUT_PATH, 'data/gaussian_output.png')
medir_tiempo('Sobel',    apply_sobel,    INPUT_PATH, 'data/sobel_output.png')
medir_tiempo('Median',   apply_median,   INPUT_PATH, 'data/median_output.png')

# --- Visualización ---
images = {
    'Original': INPUT_PATH,
    f'Gaussian ({tiempos["Gaussian"]:.2f}s)': 'data/gaussian_output.png',
    f'Sobel ({tiempos["Sobel"]:.2f}s)':    'data/sobel_output.png',
    f'Median ({tiempos["Median"]:.2f}s)':   'data/median_output.png',
}

fig, axes = plt.subplots(1, 4, figsize=(18, 5))

for ax, (title, path) in zip(axes, images.items()):
    img = Image.open(path)
    ax.imshow(img, cmap='gray' if 'Original' not in title else None)
    ax.set_title(title, fontsize=11)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
